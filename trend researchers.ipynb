```python
import requests

url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=IBM&apikey=JB01QMIO6Y6OVRTD'
r = requests.get(url)
jdata = r.json()
print(jdata)
```

```python
lDate = []
lValue = []
mts= jdata['Monthly Time Series']
for d in mts:
    dDict = mts[d]
    openV = dDict['1. open']
    lDate.append(d)
    lValue.append(openV)
print(lDate,lValue)
```

```python
import numpy as np
import matplotlib.dates as mdates
dates=[]
for d in lDate:
    date = mdates.date2num(np.datetime64(d))
    dates.append(date)
print(dates)
```

```python
import matplotlib.pyplot as plt

plt.plot_date(dates ,lValue, fmt='o', tz=None, xdate=True, ydate=False, data=None)


plt.show()
```
